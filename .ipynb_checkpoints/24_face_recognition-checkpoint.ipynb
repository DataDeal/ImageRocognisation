```python
pca_100 = PCA(n_components = 100)

faces_pca_reduced = pca_100.fit_transform(faces.data)
faces_pca_recovered = pca_100.inverse_transform(faces_pca_reduced)

face_pca_100 = faces_pca_recovered[0,:].reshape([50,37])
plt.imshow(face_pca_100, cmap = "bone")
plt.title("compressed image with 100 components", fontsize = 16)
```

```python
np.cumsum(pca_100.explained_variance_ratio_ * 100)[-1]
```

Hint: try to understand what the components represent with respect to the images. To do so, plow the components of the image 0 to 10

```python
for i in range (0,11):
    face_pca_100 = faces_pca_recovered[i,:].reshape([50,37])
    plt.imshow(face_pca_100, cmap = "bone")
    plt.title("compressed image with 100 components", fontsize = 16)
    plt.show()
```

Hint: build your classification model using your best reduction model as preprocessing in a pipeline ... use a Support Vector Machine

```python
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform

clf = make_pipeline(PCA(random_state = 0), SVC(random_state = 0))
```

```python
clf
```

Hint: perform model tuning to find the best hyper parameters. First, fine tune the SVC to reduce the complex calculation and time. Then, fine tune the PCA using a pipeline. 

```python
param_distribution = {"pca_n_components": np.arange(0,200),
                     "svc_C": reciprocal(0.1,100),
                     "svc_gamma": uniform(0.1,10),
                     "svc_kernel":["linear", "rbf", "poly"]}

rdm_search = RandomizedSearchCV(clf, param_distribution, n_iter = 10, cv = 5, random_state = 0)
```

```python
X_train, y_train = faces.data[:500], faces.target[:500]
```

```python
print(rdm_search)
```
